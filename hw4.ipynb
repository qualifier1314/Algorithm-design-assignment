## 问题 3

**道路网（Road Network）**

假设有一个以图 $ G(V, E, l) $ 表示的道路网络，连接了一组城市 $ V $。我们假设该网络是有向的，并且每条道路 $(u, v) \in E$ 都有一个非负的长度 $ l(u, v) $。一条新的道路即将被建造，因此有一个列表 $ E' $ 包含它可以连接的城市对。每对 $(u, v) \in E'$ 都有一个对应的长度 $ l'(u, v) $。我们希望选择一对城市，使得两个城市 $ s, t \in V $ 之间的距离减少最大。请为此问题编写一个高效的算法，并详细解释算法的正确性和复杂度。


```python
# idea

# 使用 Dijkstra 算法分别从 s 和 t 计算原图的最短路径
# 遍历每条候选边 (u, v) ∈ E′，评估以下三种路径替代情况是否更优
# 方案1：s -> u -> v  -> t：dist_from_s[u] + l′(u,v) + dist_to_t[v]
# 方案2：s  -> v  -> u  -> t：dist_from_s[v] + l′(v,u) + dist_to_t[u]
# 与原最短路径 d0 = dist_from_s[t] 比较，若更小则记录路径提升值 Δ = d0 − dnew
# 返回提升值最大的候选边

# code(大模型辅助完成)
import heapq
from collections import defaultdict

def dijkstra(graph, all_nodes, start):
    dist = {v: float('inf') for v in all_nodes}
    dist[start] = 0
    pq = [(0, start)]
    while pq:
        d, u = heapq.heappop(pq)
        if d > dist[u]:
            continue
        for v, w in graph[u]:
            if dist[u] + w < dist[v]:
                dist[v] = dist[u] + w
                heapq.heappush(pq, (dist[v], v))
    return dist

def reverse_graph(graph):
    rev = defaultdict(list)
    for u in graph:
        for v, w in graph[u]:
            rev[v].append((u, w))
    return rev

def find_best_edge_to_improve(V, E, E_prime, s, t):
    graph = defaultdict(list)
    for u, v, w in E:
        graph[u].append((v, w))

    dist_from_s = dijkstra(graph, V, s)
    rev_graph = reverse_graph(graph)
    dist_to_t = dijkstra(rev_graph, V, t)
    d0 = dist_from_s[t]

    best_delta = 0
    best_edge = None

    for u, v, w in E_prime:
        alt1 = dist_from_s.get(u, float('inf')) + w + dist_to_t.get(v, float('inf'))
        alt2 = dist_from_s.get(v, float('inf')) + w + dist_to_t.get(u, float('inf'))
        d1 = min(alt1, alt2)
        if d1 < d0 and (d0 - d1 > best_delta):
            best_delta = d0 - d1
            best_edge = (u, v, w)

    return best_edge, best_delta


V = ["A", "B", "C", "D", "E"]
E = [("A", "B", 4),("B", "C", 2),("C", "D", 3),("D", "E", 5),("A", "C", 10)]
E_prime = [("B", "D", 1),  ("A", "E", 15),("C", "E", 2)]
s, t = "A", "E"
best_edge, gain = find_best_edge_to_improve(V, E, E_prime, s, t)
graph_for_test = defaultdict(list)
for u, v, w in E:
    graph_for_test[u].append((v, w))
print(f"原始最短路径长度: {dijkstra(graph_for_test, V, s)[t]}")
print("最佳新增边:", best_edge)
print("最短路径可减少的距离:", gain)
```

**逃离问题**

一个 $ n \times n $ 的网格是一个无向图，由 $ n $ 行和 $ n $ 列的顶点组成，如下图所示。我们用 $(i,j)$ 表示第 $ i $ 行和第 $ j $ 列的顶点。除了边界顶点，网格中的所有顶点都有四个邻居，即满足 $ i = 1, i = n, j = 1 $ 或 $ j = n $ 的点 $(i,j)$。

给定网格中的 $ m \leq n^2 $ 个起点 $(x_1, y_1), (x_2, y_2), \cdots , (x_m, y_m)$，逃离问题是确定是否存在 $ m $ 条顶点不相交的路径（即路径之间不相交），从这些起点到边界上的任意 $ m $ 个不同点。例如，图1中的网格存在逃离。

(1) 该问题可以看作是一个最大流问题。考虑一个流网络，其中顶点和边都有容量。也就是说，进入任何给定顶点的总正流量受到容量限制。证明在具有边和顶点容量的网络中确定最大流可以简化为在具有可比大小的普通流网络上的最大流问题。更准确地说，你需要将一个具有顶点和边容量的网络 $ G = (V,E) $ 转换为另一个仅具有边容量的网络 $ G' = (V', E') $，使得两个网络上的最大流相同，并且你构建的新网络具有 $ V' = O(V) $ 个顶点和 $ E' = O(E) $ 条边。你可以假设网络是连通的。

idea：   
逃离问题建模为带顶点容量限制的最大流问题，通过将每个网格点拆分为“入点”和“出点”，并限制其间容量为1，   
从而确保路径之间顶点不重叠；再通过添加超级源点连接所有起点，超级汇点连接所有边界点，   
构建一个仅含边容量的网络，最终在该网络上运行最大流算法，判断最大流是否大于等于起点数，以确定是否存在 m 条不相交的逃离路径。

```python
#（第二问）code（大模型辅助完成）
from collections import deque, defaultdict

class Edge:
    def __init__(self, to, rev, capacity):
        self.to = to
        self.rev = rev
        self.capacity = capacity

class MaxFlow:
    def __init__(self, N):
        self.size = N
        self.graph = [[] for _ in range(N)]

    def add(self, fr, to, cap):
        forward = Edge(to, len(self.graph[to]), cap)
        backward = Edge(fr, len(self.graph[fr]), 0)
        self.graph[fr].append(forward)
        self.graph[to].append(backward)

    def bfs_level(self, s, t, level):
        queue = deque([s])
        level[s] = 0
        while queue:
            v = queue.popleft()
            for e in self.graph[v]:
                if e.capacity > 0 and level[e.to] < 0:
                    level[e.to] = level[v] + 1
                    queue.append(e.to)
        return level[t] != -1

    def dfs_flow(self, v, t, upTo, iter, level):
        if v == t:
            return upTo
        for i in range(iter[v], len(self.graph[v])):
            e = self.graph[v][i]
            if e.capacity > 0 and level[v] < level[e.to]:
                d = self.dfs_flow(e.to, t, min(upTo, e.capacity), iter, level)
                if d > 0:
                    e.capacity -= d
                    self.graph[e.to][e.rev].capacity += d
                    return d
            iter[v] += 1
        return 0

    def max_flow(self, s, t):
        flow = 0
        INF = float('inf')
        level = [-1] * self.size
        iter = [0] * self.size
        while self.bfs_level(s, t, level):
            iter = [0] * self.size
            while True:
                f = self.dfs_flow(s, t, INF, iter, level)
                if f == 0:
                    break
                flow += f
            level = [-1] * self.size
        return flow

def escape_grid(n, starts):
    def node_in(i, j): return (i * n + j) * 2
    def node_out(i, j): return (i * n + j) * 2 + 1

    S = n * n * 2  
    T = S + 1      
    mf = MaxFlow(T + 1)

    for i in range(n):
        for j in range(n):
            mf.add(node_in(i, j), node_out(i, j), 1)
            for dx, dy in [(-1,0),(1,0),(0,-1),(0,1)]:
                ni, nj = i + dx, j + dy
                if 0 <= ni < n and 0 <= nj < n:
                    mf.add(node_out(i, j), node_in(ni, nj), 1)

    for x, y in starts:
        mf.add(S, node_in(x, y), 1)

    for i in range(n):
        for j in range(n):
            if i == 0 or i == n-1 or j == 0 or j == n-1:
                mf.add(node_out(i, j), T, 1)

    flow = mf.max_flow(S, T)
    return flow == len(starts)

n = 4
starts = [(1, 1), (2, 1), (1, 2)]
print("是否存在逃离路径:", escape_grid(n, starts))
```
